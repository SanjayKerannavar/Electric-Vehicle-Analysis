# Step 3 : KPI 1 Total Vehicles (Overall, BEV, and PHEV)

```python
bev_count = dataset[dataset['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)'].shape[0]
phev_count = dataset[dataset['Electric Vehicle Type'] == 'Plug-in Hybrid Electric Vehicle (PHEV)'].shape[0]
```

```python
labels = ['Total Vehicles', 'BEVs', 'PHEVs']
```

```python
values = [total_vehicles, bev_count, phev_count]
```

```python
plt.figure(figsize=(5,4))
sns.barplot(x=labels, y=values)
plt.title('Total Vehicles vs BEVs and PHEVs')
plt.ylabel('Count')
plt.savefig('Total Vehicles vs BEVs and PHEVs.jpg')
plt.savefig('total_vehicles_bevs_phevs.png', format='png', dpi=300) 
plt.show()
```

# Step 4 : KPI 2 Average Electric Range

```python
range_dataset = dataset[dataset['Electric Range'].notnull()]
average_range = range_dataset['Electric Range'].mean()
```

```python
plt.figure(figsize=(5,4))
sns.histplot(range_dataset['Electric Range'], bins=20, kde=True)
plt.axvline(average_range, color='red', linestyle='--', label=f'Avg Range: {average_range:.2f}')
plt.title('Distribution of Electric Vehicle Range')
plt.xlabel('Electric Range (Kilometers)')
plt.legend()
plt.show()
plt.savefig('Distribution of Electric Vehicle Range.jpg')
plt.savefig('Distribution of Electric Vehicle Range.png', format='png', dpi=300) 
```

# Step 5 : KPI 3 Total Vehicles by Model Year

```python
yearly_vehicle_count = dataset[dataset['Model Year'] >= 2010].groupby('Model Year').size()
```

```python
plt.figure(figsize=(5,6))
yearly_vehicle_count.plot(kind='line')
plt.title('Total Vehicles by Model Year (From 2010)')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.show()
plt.savefig('Total Vehicles by Model Year (From 2010).jpg')
```

# Step 6 : KPI 4 Total Vehicles by State

```python
statewise_vehicle_count = dataset.groupby('State').size()
```

```python
plt.figure(figsize=(10,8))
statewise_vehicle_count.sort_values(ascending=True).plot(kind='barh')
plt.title('Total Vehicles by State')
plt.xlabel('Number of Vehicles')
plt.show(4)
```

# Step 7 : KPI 5 Top 10 Vehicles by Make

```python
top_makes = dataset['Make'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
top_makes.plot(kind='bar')
plt.title('Top 10 Makes by Total Vehicles')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

# Step 8 : KPI 6 CAFV Eligibility

```python
cafv_count = dataset['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
```

```python
plt.figure(figsize=(7,7))
cafv_count.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title('CAFV Eligibility')
plt.show()
```

# Step 10 : Conclusion Graph (Multi-panel Summary)

```python
# Create subplots for conclusion summary
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Total BEV and PHEV Vehicles
axs[0, 0].bar(['Total Vehicles', 'BEVs', 'PHEVs'], [total_vehicles, bev_count, phev_count], color=['blue', 'green', 'orange'])
axs[0, 0].set_title('Total Vehicles vs BEVs and PHEVs')
axs[0, 0].set_ylabel('Count')

# Total Vehicles by State (Top 5 States)
top_states = dataset['State'].value_counts().head(5)
axs[0, 1].bar(top_states.index, top_states.values, color='lightblue')
axs[0, 1].set_title('Top 5 States by Total Vehicles')
axs[0, 1].set_ylabel('Number of Vehicles')

# Top 10 Vehicle Makes
top_makes.plot(kind='bar', ax=axs[1, 0], color='lightgreen')
axs[1, 0].set_title('Top 10 Vehicle Makes')
axs[1, 0].set_ylabel('Number of Vehicles')

# Trend Over Time (Model Year)
yearly_vehicle_count.plot(kind='line', ax=axs[1, 1], color='red')
axs[1, 1].set_title('Total Vehicles by Model Year (From 2010)')
axs[1, 1].set_xlabel('Model Year')
axs[1, 1].set_ylabel('Number of Vehicles')

# Adjust layout
plt.tight_layout()
plt.show()

plt.savefig('ALL Vehicles.jpg')
```
